# Sistem Persamaan Linier

Persamaan linear adalah dasar dalam aljabar dan digunakan dalam berbagai bidang seperti fisika, ekonomi, dan teknik. Memahaminya membantu dalam menyelesaikan banyak masalah kehidupan nyata, seperti perhitungan keuangan, kecepatan, dan optimasi.

### Tugas

### Penyelesaian Sistem Persamaan Linear 3 Dimensi Menggunakan Eliminasi Gauss
Metode eliminasi Gauss digunakan untuk menentukan penyelesaian sistem persamaan linear, baik yang memiliki satu solusi, tak hingga solusi, maupun tanpa solusi.

---

### 1. **Satu Solusi (Unik)**  
Contoh sistem persamaan:

x + y + z = 6

2x + 3y + z = 14

y + 2z = 8


**Matriks Augmented:**


1 & 1 & 1 & | & 6 
2 & 3 & 1 & | & 14 
0 & 1 & 2 & | & 8


Langkah-langkah:

1. Tukar baris jika perlu.
2. Eliminasi elemen di bawah elemen utama (pivot).
3. Bentuk eselon baris.
4. Substitusi mundur.

Hasil akhirnya:


x = 2, \quad y = 4, \quad z = 2

**Kesimpulan:** Sistem memiliki **satu solusi unik**.

Link:
https://www.geogebra.org/m/j7bz5sz9

---

### 2. **Tak Hingga Solusi**
Contoh sistem persamaan:

x + y + z = 5

2x + 2y + 2z = 10

3x + 3y + 3z = 15

**Matriks Augmented:**


1 & 1 & 1 & | & 5 
2 & 2 & 2 & | & 10 
3 & 3 & 3 & | & 15

Bentuk eselon baris:

1 & 1 & 1 & | & 5 
0 & 0 & 0 & | & 0 
0 & 0 & 0 & | & 0


**Kesimpulan:** Sistem memiliki **tak hingga banyak solusi** karena ada baris nol yang menunjukkan persamaan redundan.

Link:
https://www.geogebra.org/m/dtxnmxwa

---

### 3. **Tanpa Solusi**
Contoh sistem persamaan:

x + y + z = 3

2x + 2y + 2z = 6

x + y + z = 5

**Matriks Augmented:**

1 & 1 & 1 & | & 3 
2 & 2 & 2 & | & 6 
1 & 1 & 1 & | & 5


Bentuk eselon baris:

1 & 1 & 1 & | & 3 
0 & 0 & 0 & | & 0 
0 & 0 & 0 & | & 2


Baris terakhir menunjukkan **kontradiksi (0 = 2)**.

**Kesimpulan:** Sistem **tidak memiliki solusi (inkonsisten)**.

https://www.geogebra.org/m/ut7hraa7
